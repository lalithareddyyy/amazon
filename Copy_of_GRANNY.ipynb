```
import tensorflow as tf
import gpt_2_simple as gpt2_simple
import os
from gtts import gTTS
from IPython.display import Audio, display

def generate_story(genre, starting_line):
    # Define the paths based on the selected genre
    if genre == 'scary':
        checkpoint_path = '/content/gpt-2/checkpoint/run2'
        story_path = '/content/drive/MyDrive/Project_Granny/Scary/scary.txt'
    elif genre == 'humour':
        checkpoint_path = '/content/gpt-2/checkpoint/run3'
        story_path = '/content/drive/MyDrive/Project_Granny/Humour/humour.txt'
    else:
        return "Invalid genre selected"

    # Fine-tune the model if the checkpoint doesn't exist
    if not os.path.exists(checkpoint_path + '/checkpoint'):
        gpt2_simple.finetune(sess, dataset=story_path, steps=100, model_name='124M',
                             sample_every=20, save_every=100, print_every=10, restore_from='fresh', run_name=checkpoint_path.split('/')[-1])

    # Load the model with the specified checkpoint path
    gpt2_simple.load_gpt2(sess, run_name=checkpoint_path, reuse=tf.compat.v1.AUTO_REUSE)

    # Generate story based on the starting line
    generated_story = gpt2_simple.generate(
        sess,
        run_name=checkpoint_path,
        prefix=starting_line,
        length=400,  # Adjust length as needed
        temperature=0.7,
        return_as_list=True
    )[0]

    # Display the generated story
    print("Generated Story:")
    print(generated_story)

    # Convert the text to speech and play the audio
    tts = gTTS(generated_story, lang='en')
    tts.save("generated_audio.mp3")
    display(Audio("generated_audio.mp3", autoplay=True))

# Example usage:
genre_input = input("Enter the genre (scary or humour): ")
starting_line_input = input("Enter the starting line of the story: ")

generate_story(genre_input, starting_line_input)
```
